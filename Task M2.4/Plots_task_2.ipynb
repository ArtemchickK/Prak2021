```python
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt
```

```python
case1_1 = [0.4, 1.2, 0.05, 0.375]
case1_2 = [0.4, 1.2, 0.05, 0.712]
case1_3 = [0.4, 1.2, 0.05, 0.7877]
```

```python
case2_1 = [2,6,0.7,3]
case2_2 = [2,6,0.8,3]
case2_3 = [2,6,0.9,3]
case2_4 = [2,6,0.6,3]
case2_5 = [2,6,0.6,3]
```

```python
case3_1 = [2,6,0.5,6]
```

```python
def system_of_ode(s,t,A,B):
    y1, y2 = s
    return A-(B)*y1-y1+y1**2*y2, B*y1 - y1**2*y2
```

```python
def get_solution(case):
    eps = 0.05
    dt = 0.01; t_stop = 250
    iters = int(t_stop/dt)
    t = np.linspace(0, t_stop, iters)
    A0,B,a,W = case
    A = A0 + a*np.cos(W)
    start_state = [A + eps, B/A + eps]
    solution = odeint(system_of_ode, start_state, t, args=(A,B))
    return solution
```

```python
def draw_y1(sol):
    solution = sol
    dt = 0.01; t_stop = 250
    iters = int(t_stop/dt)
    t = np.linspace(0, t_stop, iters)
    plt.plot(t, solution[:, 0], 'r')
    plt.xlim([0,250])
```

```python
def draw_y2(sol):
    solution = sol
    dt = 0.01; t_stop = 250
    iters = int(t_stop/dt)
    t = np.linspace(0, t_stop, iters)
    plt.plot(t, solution[:, 1])
    plt.xlim([0,150])
```

```python
def phase_plane(sol):
    solution = sol
    plt.plot(solution[:,0], solution[:, 1],lw = 0.5)
```

(1) $A_{0} = 0.4, B = 1.2, a = 0.05, W = 0.375 $

```python
solution1_1 = get_solution(case1_1)
draw_y1(solution1_1)
```

```python
draw_y2(solution1_1)
```

```python
phase_plane(solution1_1)
```
